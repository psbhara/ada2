```python
costs = [2, 3, 4, 5]  
profits = [3, 4, 5, 6] 
budget = 5             
```

```python


def knapsack(costs, profits, budget):
    n = len(profits)
    
    
    dp = [[0] * (budget + 1) for _ in range(n + 1)]
    
    for i in range(1, n+1):          
        for w in range(1, budget+1): 
            if costs[i-1] <= w:
                dp[i][w] = max(dp[i-1][w], dp[i-1][w-costs[i-1]] + profits[i-1])
            else:
                dp[i][w] = dp[i-1][w]
    
    return dp[n][budget], dp

```

```python
costs = [2, 3, 4, 5]
profits = [3, 4, 5, 6]
budget = 5

max_profit, dp_table = knapsack(costs, profits, budget)
print("Maximum Profit:", max_profit)
```

```python
import matplotlib.pyplot as plt

budgets = list(range(1, budget+1))
profits_vs_budget = [dp_table[len(costs)][b] for b in budgets]

plt.figure(figsize=(6,4))
plt.plot(budgets, profits_vs_budget, marker='o', linestyle='--', color='g', label="Profit Growth")
plt.title("Profit vs Budget")
plt.xlabel("Budget")
plt.ylabel("Maximum Profit")
plt.legend()
plt.grid(True)
plt.show()
```
